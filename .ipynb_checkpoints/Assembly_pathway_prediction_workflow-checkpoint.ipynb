## 4. Implementation of assembly pathway preditction workflow

<a id="section_ID_4.2"></a>
### 4.1 Method to process the graph to find all labels

```python
def get_labels(complex_graph):
    """
    Finds all labels for the vertices

    :param complex_graph: graph file in gml format
    :returns: list with labels
    """
    nodes = complex_graph.vs
    labels = nodes["label"]
    return labels
```

<a id="section_ID_4.3"></a>
### 4.2 Methods for newick string conversion

```python
def get_newick(node, parent_dist, leaf_names):
    """
    Convert sciply.cluster.hierarchy.to_tree()-output to Newick format.

    :param node: output of sciply.cluster.hierarchy.to_tree()
    :param parent_dist: output of sciply.cluster.hierarchy.to_tree().dist
    :param leaf_names: list of leaf names
    :returns: tree in Newick format
    """
    if node.is_leaf():
        return leaf_names[node.id]
    left_newick = get_newick(node.get_left(), node.dist, leaf_names)
    right_newick = get_newick(node.get_right(), node.dist, leaf_names)
    return f"({left_newick},{right_newick}):{parent_dist}"


def newick_parameters(linkage_matrix, labels):
    """ 
    Convert a given linkage matrix from scipy to a newick string
    
    :param: linkage_matrix
    :returns: linkage matrix as newick string 
    """
    tree = sch.to_tree(linkage_matrix)
    newick = get_newick(tree, tree.dist, labels)
    return f"({newick});"
```

<a id="section_ID_4.4"></a>
### 4.3 Method for creation of distance matrix

```python
def create_distance_matrix(complex_graph):
    """ 
    Reads in the complex graph and returns the condensed distance matrix 
    
    :param: complex_graph: graph saved as object from igraph libary 
    :returns: condensed_matrix 
    """
    adj_matrix_absw = complex_graph.get_adjacency(attribute="absoluteWeight")
    adj_array_absw_list = np.array(adj_matrix_absw.data)
    # Calculates the distance matrix
    max_value = np.max(adj_array_absw_list)
    dist_matrix = -1 * (adj_array_absw_list - max_value) / max_value
    np.fill_diagonal(dist_matrix, 0)
    # Calculates the condensed distance matrix
    condensed_matrix = squareform(dist_matrix)
    return condensed_matrix
    
```

<a id="section_ID_4.5"></a>
### 4.4 Cluster algorithms

<a id="section_ID_4.5.1"></a>
#### 4.4.1 Ward linkage Clustering 

```python
def ward_linkage_clustering(condensed_matrix):
    """ 
    Reads in the condensed distance matrix and creates tthe tre clustered with ward linkage
    
    :param: condensed_matrix: condensed distance matrix saved as numpy array 
    :returns: ns_ward_absw: contains the tree as newick string  
    """
    ward_linkage_matrix = sch.linkage(condensed_matrix, method="ward", metric='euclidean', optimal_ordering=True)

    fig = plt.figure(figsize=(10, 5))
    fig.suptitle("absolute weight ward linkage", fontsize=16)
    dn = sch.dendrogram(ward_linkage_matrix, labels = labels)
    ns_ward_absw = newick_parameters(ward_linkage_matrix, labels)
    return ns_ward_absw
```

#### <a id="section_ID_4.5.2"></a>
#### 4.4.2 Weighted linkage Clustering

```python
def weighted_linkage_clustering(condensed_matrix):
    """ 
    Reads in the condensed distance matrix and creates the tree clusterd with weighted linkage 
    
    :param: condensed_matrix: condensed distance matrix saved as numpy array 
    :returns: ns_ward_absw: contains the tree as newick string  
    """
    weighted_linkage_matrix = sch.linkage(condensed_matrix, method="weighted", metric='euclidean', optimal_ordering=True)
    
    fig= plt.figure(figsize=(10, 5))
    #plt.grid(axis='y')
    #plt.yticks(np.arange(0, 1.1, 0.05))
    fig.suptitle("absolute weighted linkage", fontsize=16)
    dn = sch.dendrogram(weighted_linkage_matrix, labels = labels)
    ns_weighted_absw = newick_parameters(weighted_linkage_matrix, labels)
    return ns_weighted_absw
```

<a id="section_ID_4.5.3"></a>
#### 4.4.3 UPGMA linkage Clustering

```python
def UPGMA_linkage_clustering(condensed_matrix):
    """ 
    Reads in the condensed distance matrix and creates the tree clustered with UPGMA
    
    :param: condensed_matrix: condensed distance matrix saved as numpy array 
    :returns: ns_ward_absw: contains the tree as newick string  
    """
    UPGMA_linkage_matrix = sch.linkage(condensed_matrix, method="average", metric='euclidean', optimal_ordering=True)

    fig = plt.figure(figsize=(10, 5))
    fig.suptitle("absolute weight UPGMA", fontsize=16)
    dn = sch.dendrogram(UPGMA_linkage_matrix, labels = labels)
    ns_UPGMA_absw = newick_parameters(UPGMA_linkage_matrix, labels)
    return ns_UPGMA_absw
```

### 5.3 Method to calculate if a tree is a subtree 

```python
# Define a custom TreeNode class to represent a tree node with a name and children
class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []

        
def convertEte3ToTreeNode(ete3_node):
    """ 
    Function to convert an ete3 node to a TreeNode
    
    :param: ete3_node: contains the ete3 node which will be converted 
    :returns: node: the ete3 node converted to a TreeNode object   
    """
    # Converts an ete3 node to a custom tree node
    node = TreeNode(ete3_node.name)
    for child in ete3_node.children:
        node.children.append(convertEte3ToTreeNode(child))
    return node


def isSubtree(tree, subtree):
    """ 
    Function to check if subtree is a subtree of tree 
    
    :param: tree: ete3 tree object which is the supposed parenttree 
    :param: subtree: ete3 tree object which is the supposed subtree   
    """
    if tree is None:
        return False
    if isIdentical(tree, subtree):  # If 'tree' and 'subtree' are identical, 'subtree' is considered a subtree
        return True
    return any(isSubtree(child, subtree) for child in tree.children)


def isIdentical(node1, node2):
    """ 
    Function to check if two noes are identical 
    
    :param: node1, node2: The two nodes which are examined 
    :returns: True: If they are identical
    :teturns: False: If they are different 
    """
    if node1 is None and node2 is None:
        return True
    if node1 is None or node2 is None:
        return False
    if node1.name != node2.name:
        return False
    if len(node1.children) != len(node2.children):
        return False
    return all(isIdentical(child1, child2) for child1, child2 in zip(node1.children, node2.children))
```

<a id="section_ID_5.4.1"></a>
#### 5.4.1 Comparision algorithm

##### This part of the code is  destinated  for publication 
##### Subtree comparision algorithm with filter for  minimum subtrees in self calculated tree

```python
def tree_comparision(reference_tree, calculated_tree, threshold):
    """ 
    Function to find all minimum subtrees more common then the threshold
    
    :param: reference_tree: The reference tree in newick format 
    :param: calculated_tree: The calculated tree in newick format
    :param: : threshold: The threshold for the norm robinson foulds distance 
    :returns: found_subtrees : Contains all minimum subtrees more common then the treshold 
    :returns: found_norm_rf : Contains all norm robinsons fould distances between the minimum subtrees and the corresponding subtree structures  in the reference  
    """
    subtrees = []  # List which holds all subtrees from self calculated tree 
    subtrees_ref = []  # List which holds all subtrees from reference tree 
    found_subtrees = []  # List which holds all subtrees found in reference and self calculated tree 
    found_norm_rf = [] # List which holds all found norm robinson foulds distances
    old_norm_rf = 2 # Old norm robinson foulds distance

    # Calculate ete3 tree from newick string 
    ete3_tree = Tree(reference_tree)
    ete3_subtree = Tree(calculated_tree)
    old_calculated_tree = None # Old subtree 

    # Calcuate all subtrees from self calculated tree 
    for node in ete3_subtree.traverse("preorder"):
        if node.children != []:
            subtrees.append(node)

    # Calculate all subtrees from reference        
    for node in ete3_tree.traverse("preorder"):
        if node.children != []:
            subtrees_ref.append(node)

    # Compare subtrees from reference with subtrees from self calculated tree 
    print("Reference tree: ")
    print(ete3_tree)
    for ref_tree in subtrees_ref:  # iterates through all subtrees of the reference 
        for calculated_tree in subtrees:  # iterates through all subtrees of the calculated tree  
            results = ref_tree.compare(calculated_tree, unrooted=True)  # compares the currently examined subtrees 
            try:
                if results["norm_rf"] < threshold:  # all comparisons with a norm robinson foulds distance under 0.4 are  relevant 
                    if old_calculated_tree == None:  # first examined tree cant have a predecessor tree 
                        old_calculated_tree = calculated_tree  # set the first tree as last tree 
                        old_norm_rf = results["norm_rf"]  # set the last norm robinson foulds distance
                    else:
                        tree = convertEte3ToTreeNode(old_calculated_tree)  # last examined tree as TreeNode object 
                        subtree = convertEte3ToTreeNode(calculated_tree)  # currently examined tree as TreeNode object 
                        resulter = isSubtree(tree, subtree)  # checks if the currently examined tree is a subtree from the last examined tree 
                        if resulter == True:   # If the currently examined tree is a subtree from the last tree dont add it to the list 
                            old_calculated_tree = calculated_tree  # change the currently examined tree to the last examined tree
                            old_norm_rf = results["norm_rf"]  # change the current norm robinson foulds distance to last norm robinson fould dinstance 
                        else:   # If the tree is not a subtree from the last tree add it to the list 
                            for i in range(len(found_subtrees)):  # checks if the substructur is annotated with the smallest norm rf possible 
                                if old_calculated_tree == found_subtrees[i] and old_norm_rf <= found_norm_rf[i]: 
                                    found_norm_rf.pop(i)
                                    found_norm_rf.insert(i, old_norm_rf)
                            if old_calculated_tree not in found_subtrees:  # if the tree is no subtree of the last tree check if the tree is already in the list of found subtrees 
                                found_subtrees.append(old_calculated_tree)  # if not append it to found subtrees
                                found_norm_rf.append(old_norm_rf)   # also append the corresponding norm robinson foulds distance 
                            old_calculated_tree = calculated_tree  # change the currently examined tree to the last examined tree 
                            old_norm_rf = results["norm_rf"]   # change the current norm robinson foulds distance to last norm robinson fould dinstance 
            except TypeError:
                pass

    for i in range(len(found_subtrees)):
        print("This substructur has been found in the self calculated tree: ")
        print("Norm Robinson Foulds Distance: ", found_norm_rf[i])
        print("Subtree structure: ")
        print(found_subtrees[i])
    return found_subtrees, found_norm_rf

```

```python
# compares the reference tree and calculated tree 
# its possible to change the input trees to compare any calculted trees and reference trees 
# its possible to change 0.4  to any other threshold 
# subtrees is the list with all minimum subtrees found to be more common then the given threshold
# norm_robinson_fould contains all corresponding calculated norm robinson fould distances
subtrees, norm_robinson_foulds = tree_comparision(Newick_ref1, UPGMA_linkage, 0.4)
```
